```python
# total_runs_y have the runs scored at each ball
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['overs_faced'] = delivery_df['over'] + (delivery_df['ball'] / 6.0)
delivery_df['current_run_rate'] = delivery_df['current_score'] / delivery_df['overs_faced']
delivery_df.columns
```

```python
final_df['batting_team'] = final_df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')

final_df['batting_team'] = final_df['batting_team'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')

final_df['batting_team'] = final_df['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')

final_df['batting_team'] = final_df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

final_df['batting_team'] = final_df['batting_team'].str.replace('Gujarat Lions','Gujarat Titans')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
groups = final_df.groupby('match_id')
match_ids = final_df['match_id'].unique()
last_five = []
for id in match_ids:
    group = groups.get_group(id)
    group['total_runs_y'] = pd.to_numeric(group['total_runs_y'], errors='coerce') # converts any error into nan
    rolling_sum = group['total_runs_y'].rolling(window=30).sum() # for calculating score of last 5 overs
    last_five.extend(rolling_sum.values.tolist())
```

```python
final_df['last_five_over']=last_five
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','current_run_rate','last_five_over','total_runs_x']]
```

```python
y=final_df['total_runs_x']
x=final_df.drop(columns=['total_runs_x'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)

print(x_test.shape,y_test.shape)
print(x_train.shape,y_train.shape)
```

```python
import pandas as pd
# Combine x_train and y_train for training data
train_data = pd.concat([x_train, y_train], axis=1)

# Combine x_test and y_test for testing data
test_data = pd.concat([x_test, y_test], axis=1)

# Save the training data to a CSV file
train_data.to_csv('D://InningsInsights(Python Individual)//datasets//train_data.csv', index=False)

# Save the testing data to a CSV file
test_data.to_csv('D://InningsInsights(Python Individual)//datasets//test_data.csv', index=False)

print("Training and testing data have been saved to 'train_data.csv' and 'test_data.csv'.")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# OneHotEncoder transforms categorical features into binary matrix
# ColumnTransformer is used to apply transformations to specific columns while leaving other columns unchanged.
# sparse_output returns the result as a NumPy array instead of a sparse matrix.
# the three column specified is in categorical form
# 'passthrough' means that any columns not included in the transformer steps will be passed through unchanged.... 
# This is useful if you have numerical columns or other types of data that should not be altered. 
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
# Simplify the workflow for data preprocessing and model training
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

What?
RandomForestClassifier is an ensemble learning algorithm that constructs multiple decision trees during training and outputs the mode (class) of the classes (classification) of the individual trees. The basic idea is to combine the predictions of several base classifiers (decision trees) to improve the robustness and accuracy of the model.

When?
Ensemble Method: Combines the predictions of multiple decision trees to make a final decision.
Bagging: Uses bootstrapped subsets of the training data to train each tree, which helps in reducing variance and improving model stability.
Feature Randomness: At each split in the decision trees, it considers a random subset of features, which helps in making the model more robust and less prone to overfitting.

Why?
Accuracy: Often achieves high accuracy compared to other classifiers due to its ensemble approach.
Versatility: Can be used for both classification and regression tasks (with RandomForestRegressor).
Handling Missing Values: Can handle missing values in the dataset without requiring imputation.
Robustness: Less likely to overfit than individual decision trees, especially when the number of trees is large.
Feature Importance: Provides feature importance scores which can be helpful for feature selection and understanding the model.
